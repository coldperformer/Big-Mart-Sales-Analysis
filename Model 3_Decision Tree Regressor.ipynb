```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
%matplotlib inline
```

```python
df = pd.read_csv('Cleaned_Dataset.csv')
df.head(10)
```

```python
encoder = LabelEncoder()
for i in df.columns:
    if isinstance(df[i][0], str):
        df[i] = encoder.fit_transform(df[i])
```

```python
df.head()
```

```python
X = df.drop(['Item_Identifier','Item_Weight','Item_Fat_Content','Item_Visibility','Outlet_Establishment_Year','Item_Outlet_Sales'], axis=1)
y = df['Item_Outlet_Sales']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=1)
```

```python
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(X_train,y_train)
print("Training Accuracy:",np.round(dtr.score(X_train,y_train), decimals=2)*100)
print("Testing Accuracy:",np.round(dtr.score(X_test,y_test), decimals=2)*100)
```

```python
train_score = []
test_score = []
for i in range(1,11):
    dtr = DecisionTreeRegressor(max_depth=i, random_state=0)
    dtr.fit(X_train,y_train)
    train_score.append(dtr.score(X_train,y_train))
    test_score.append(dtr.score(X_test,y_test))
```

```python
max_depth = range(1,11)
fig = plt.figure(figsize=[13.66,7.68])
plt.plot(max_depth, train_score)
plt.plot(max_depth, test_score)
plt.grid()
plt.legend(['Train','Test'], loc='upper right')
plt.xlabel('Max Depth')
plt.ylabel('Score')
```

```python
train_score = []
test_score = []
for i in range(2,21):
    dtr = DecisionTreeRegressor(max_leaf_nodes=i, random_state=0)
    dtr.fit(X_train,y_train)
    train_score.append(dtr.score(X_train,y_train))
    test_score.append(dtr.score(X_test,y_test))
```

```python
max_leaf_nodes = range(2,21)
fig = plt.figure(figsize=[13.66,7.68])
plt.plot(max_leaf_nodes, train_score)
plt.plot(max_leaf_nodes, test_score)
plt.grid()
plt.legend(['Train','Test'], loc='upper right')
plt.xlabel('Max Leaf Nodes')
plt.ylabel('Score')
```

```python
dtr = DecisionTreeRegressor(criterion='mse',splitter='best',max_depth=5,max_leaf_nodes=15,random_state=0,min_samples_leaf=100,)
dtr.fit(X_train,y_train)
print("Training Accuracy:",np.round(dtr.score(X_train,y_train), decimals=2)*100)
print("Testing Accuracy:",np.round(dtr.score(X_test,y_test), decimals=2)*100)
```
