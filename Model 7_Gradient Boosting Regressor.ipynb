```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
%matplotlib inline
```

```python
df = pd.read_csv('Cleaned_Dataset.csv')
df.head()
```

```python
encoder = LabelEncoder()
for i in df.columns:
    if isinstance(df[i][0], str):
        df[i] = encoder.fit_transform(df[i])
df.head()
```

```python
X = df.drop(['Item_Identifier','Item_Weight','Item_Fat_Content','Item_Visibility','Outlet_Establishment_Year','Item_Outlet_Sales'], axis=1)
y = df['Item_Outlet_Sales']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=0)
```

```python
gbr = GradientBoostingRegressor(loss='ls', learning_rate=0.025, n_estimators=100,subsample=1.0,criterion='friedman_mse',max_depth=3,alpha=0.9)
gbr.fit(X_train, y_train)
print('Train Accuracy:',np.round(gbr.score(X_train, y_train), decimals=2)*100)
print('Test Accuracy:',np.round(gbr.score(X_test, y_test), decimals=2)*100)
```
